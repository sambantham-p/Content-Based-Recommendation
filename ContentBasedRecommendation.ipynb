```python
import pandas as pd
import re
from google.colab import drive
drive.mount('/content/drive')
train_ds = pd.read_csv('/content/drive/My Drive/train.csv')
train_ds.head()
```

```python
train_ds[train_ds.isnull().any(axis=1)]
```

```python
product = []
for i in train_ds.title.unique():
  product.append(str(i))
```

```python
len(product)
product1 = train_ds.copy()
for category in product:
    product1[category] = 0
product1['title'] = product1['title'].fillna('')
for title in product:
    title = re.escape(title)
    product1[title] = product1['title'].str.contains(title).astype(int)
product1.drop(columns=['title'], inplace=True)
product1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
data = product1.copy()
title = data.columns.drop(['id','platform','Rating','maincateg','price1','actprice1','Offer %','norating1','noreviews1','star_5f','star_4f','star_3f','star_2f','star_1f','fulfilled1'])
sum_labels = data[title].sum()
product_dv = pd.DataFrame({'Labels':title, 'Total': sum_labels})
f, ax = plt.subplots(1, 1, figsize=(60, 15))
sns.barplot(data=product_dv, x='Labels', y='Total', axes=ax)
ax.set(ylim=(1, 150), ylabel='Number of products')
plt.xticks(rotation=90)
plt.show()
```
